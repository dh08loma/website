```python
def initialize_dict(d):
    d={'Off_Eff_FGperc':[],
            'Def_Eff_FGperc':[],
            'Off_ORRate':[],
            'Def_ORRate':[],
            'Off_TORate':[],
            'Def_TORate':[],
            'Off_FTRate':[],
            'Def_FTRate':[],
            'wins_82':[],
            'Color':[]}
    
    return d
    
```

```python
for i in data.index.values:
    x=data.ix[i]
    if x.Team == 'New Jersey Nets':
        data.ix[i,'Team']='Brooklyn Nets'
    if x.Team == 'Vancouver Grizzlies':
        data.ix[i,'Team']='Memphis Grizzlies'
    if x.Team == 'Seattle SuperSonics':
        data.ix[i,'Team']='Oklahoma City Thunder' 
    if x.Team == 'Charlotte Hornets' and x.Year < 2014:
        data.ix[i,'Team']='New Orleans Pelicans'
    if x.Team == 'New Orleans Hornets':
        data.ix[i,'Team']='New Orleans Pelicans'
    if x.Team == 'Charlotte Bobcats':
        data.ix[i,'Team']='Charlotte Hornets'
```

```python
def get_color(x):
    team=x.Team
    if team in ['Washington Wizards','Dallas Mavericks','Philadelphia 76ers','Oklahoma City Thunder']:
        return 'blue'
    if team == 'Boston Celtics':
        return 'green'
    if team in ['Houston Rockets','Miami Heat','Toronto Raptors','Chicago Bulls','Portland Trailblazers','Los Angeles Clippers']:
        return 'red'
    if team in ['Los Angeles Lakers','Sacramento Kings']:
        return 'purple'
    if team == 'Charlotte Hornets':
        return 'teal'
    if team in ['San Antonio Spurs','Brooklyn Nets']:
        return 'black'
    if team in ['Phoenix Suns','Utah Jazz','New York Knicks']:
        return 'orange'
    if team in ['Cleveland Cavaliers','Atlanta Hawks']:
        return 'maroon'
    if team == 'Denver Nuggets':
        return 'lightblue'
    if team in ['Golden State Warriors','Indiana Pacers']:
        return 'gold'
    if team in ['Memphis Grizzlies','Orlando Magic','Minnesota Timberwolves']:
        return 'steelblue'
    if team == 'Milwaukee Bucks':
        return 'darkgreen'
    if team in ['New Orleans Pelicans','Detroit Pistons']:
        return 'darkblue'
    
data['Color']=data.apply(lambda x: get_color(x),axis=1)
```

```python
def normalize(x,data,stat):
    data_year=data[data.Year==x.Year]
    return (x[stat]-np.mean(data_year[stat]))/np.std(data_year[stat])

def normalize_inv(x,data,stat):
    data_year=data[data.Year==x.Year]
    return (x[stat]-np.mean(data_year[stat]))/np.std(data_year[stat])*-1

for i in [2,3,5,8]:
    data[data.columns[i]]=data.apply(lambda x: normalize(x,data,data.columns[i]),axis=1)
    
for i in [4,6,7,9]:
    data[data.columns[i]]=data.apply(lambda x: normalize_inv(x,data,data.columns[i]),axis=1)
    
data
```

```python
def get_perc(x,data,stat):
    return scipy.stats.norm.cdf(x[stat])


for i in [2,3,5,8]:
    data[data.columns[i]]=data.apply(lambda x: get_perc(x,data,data.columns[i]),axis=1)
    
for i in [4,6,7,9]:
    data[data.columns[i]]=data.apply(lambda x: get_perc(x,data,data.columns[i]),axis=1)
    
data
```

```python
stats_dict={'Off_Eff_FGperc':[],
            'Def_Eff_FGperc':[],
            'Off_ORRate':[],
            'Def_ORRate':[],
            'Off_TORate':[],
            'Def_TORate':[],
            'Off_FTRate':[],
            'Def_FTRate':[],
            'wins_82':[],
            'Color':[]}

team_dict={}

for team in data.Team.unique():
    stats_dict=initialize_dict(stats_dict)
    team_stats=data[data.Team==team]
    for year in team_stats.Year:
        data_year=team_stats[team_stats.Year==year]
        for stat in stats_dict:
            if stat=='wins_82':
                stats_dict[stat].append([year,int(data_year[stat])])
            elif stat=='Color':
                stats_dict[stat].append([year,np.array(data_year.Color)[0]])
            else:
                stats_dict[stat].append([year,float(data_year[stat])])
                
    team_dict[team]=stats_dict
```

```python
stat_map={'Off_Eff_FGperc':'Offensive EFg%','Off_ORRate':'Offensive Off. Reb. %','Off_TORate':'Offensive TO Rate',
          'Off_FTRate':'Offensive FT Rate','Def_Eff_FGperc':'Defensive EFg%','Def_ORRate':'Defensive Off. Reb. %','Def_TORate':'Defensive TO Rate',
          'Def_FTRate':'Defensive FT Rate'}
```

```python
relevant_stats=['Off_Eff_FGperc', 'Off_ORRate', 'Off_TORate', 'Off_FTRate', 'Def_Eff_FGperc', 'Def_ORRate', 'Def_TORate', 'Def_FTRate']
spider_dict={}

team_groups=data.groupby('Team',axis=0)
for name, group in team_groups:
    spider_dict[name]={}
    year_groups=group.groupby('Year',axis=0)
    for year,group1 in year_groups:
        spider_dict[name][year]=[]
        for stat in relevant_stats:
            spider_dict[name][year].append({'axis':stat_map[stat],'value':round(float(group1[stat]),4)})
```

```python
import json
with open(filepath+'spider_data.json', 'w') as outfile:
    json.dump(spider_dict, outfile)
```
